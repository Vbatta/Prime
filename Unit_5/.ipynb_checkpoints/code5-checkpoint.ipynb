```python
# convert json to python object
import json
json_str='{"name":"Vaibhav","isTeacher":true,"value":null}'
py_obj = json.loads(json_str)
print(py_obj)
```

```python
# convert python to json object
py_obj={"name":"Vaibhav","isTeacher":True,'value':None}
json_obj = json.dumps(py_obj)
print(json_obj)
```
