```python
#1
str1 ="raececar"
i=0
flag = False
for ch in str1:
    if(str1[i]==(str1[len(str1)-i-1])):
        print(str1[i])
        flag = True
    else:
        flag = False
        break
    i+=1
if(flag ==True):
    print("str is palindrome number")
        
```

```python
# Optimized version
str1 ="raececar"
i=0
flag = True
for ch in str1:
    if(str1[i]!=(str1[len(str1)-i-1])):
        flag = False
        break
    i+=1 # remember
if(flag == True):
    print("it's a palindrome no.")
else:
    print("it's not a palindrome no.")
```

```python
# Logic-based
lst1 =[1,5,3]
lst2 =[4,2,7]
merged =lst1+lst2
print(merged)

for i in range(len(merged)):
    for j in range(i+1,len(merged)):
        if(merged[i]>merged[j]):
            var =merged[j]
            merged[j]=merged[i]
            merged[i] = var
            
print(merged)
          
```

```python
#4
tup1 =(1,2,3,4,5,6,6,7,8)
lst1=[]
lst2=[]
for i in tup1:
    if(tup1[i]%2==0):
        lst1.append(tup1[i])
    else:
        lst2.append(tup1[i])

tup1 =tuple(lst1)
tup2 =tuple(lst2)

print(tup1)
print(tup2)
```

```python
# Using tuple comprehension
tup1 =(1,2,3,4,5,6,6,7,8)

tup2 =tuple(i for i in tup1 if i%2 == 0 )
print(tup2)

tup3 =tuple(i for i in tup1 if i%2 != 0 )
print(tup3)
```

```python
#5
dict1={"Aman":98,"Nisha":87}
var = input("enter operation :'A'/'B'/'C'/'D' ").upper()
match var:
    case 'A':
        print('<--- Add a student --->')
        name =input("enter name :")
        marks =int(input("enter marks :"))
        dict1[name] = marks
    case 'B':
        print('<--- Update marks --->')
        name =input("enter name :")
        if (dict1.get(name) !=None):
            marks =int(input("enter marks :"))
            dict1[name]=marks
        else:
            print("user doesn't exists in dictionary")    
    case 'C':
        print('<--- Search for a student --->')
        name =input("enter name :")
        if name in dict1.keys():
            print(f"{name} exists in dictionary")
        else:
            print("user doesn't exists in dictionary") 
        
    case 'D':
        print("<--- Display all student and marks --->")
        for name,marks in dict1.items():
            print(f"{name}:{marks}")
    case _:
        print("Wrong key entered")

print(dict1)
```

```python
# 8
# share no common elements list1 = [1, 2, 3, 4] list2 = [5, 6, 7, 8]
# share common elements list1 = [1, 2, 3] list2 = [3, 4]
list1 = [1, 2, 3, 4, 7] 
list2 = [5, 6, 7, 8]
unique =set()
flag = True
for i in list1:
    unique.add(i)

for i in list2:
    if i not in unique:
        unique.add(i)

    else:
        flag =False
        break
print(unique)
if flag:        
    print("Share no common elements")
else:
    print("Share common elements")
```

```python
# optimized logic
list1 = [1, 2, 3, 4, 7]
list2 = [5, 6, 7, 8]

unique = set(list1)
flag = True

for i in list2:
    if i in unique:
        flag = False
        break

if flag:
    print("Share no common elements")
else:
    print("Share common elements")
```

```python
# 9 
list1 = [1, 2, 3, 4, 7, 3, 4]
set1 =set()
set2 =set()

for i in list1:
    if i not in set1:
        set1.add(i)
    else:
        set2.add(i)
    
print(set2)
    
    
```

```python
# Better variable names
list1 = [1, 2, 3, 4, 7, 3, 4]

seen = set()
duplicates = set()

for i in list1:
    if i in seen:
        duplicates.add(i)
    else:
        seen.add(i)

print("Elements appearing more than once:", duplicates)
```
