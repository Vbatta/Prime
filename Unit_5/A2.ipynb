```python
# optimized
def print_digits(n):
    digits = []

    while n > 0:
        digits.append(n % 10)
        n = n // 10   # integer division

    for d in reversed(digits):
        print(d)

print_digits(312)
```

```python
#4
def count_ret(n):
    count =0
    while n>0:
        n = n //10
        count +=1
    return count

print(count_ret(1234))
```

```python
#5
def sum_ret(n):
    sum_ =0
    while n>0:
        a = n %10
        sum_+=a
        n = n//10
    return sum_

print(sum_ret(1234))
```

```python
#8
def calculator(a,b,op):
    match op:
        case "+":
            print(a+b)
        case "-":
            print(a-b)
        case "*":
            print(a*b)
        case "/":
            print(a/b)
    
calculator(1,2,'/')
        
```

```python
# 9 corrected
def is_Prime(n):
    if n <= 1:
        print("not prime")
        return

    flag = False
    for i in range(2, n):
        if n % i == 0:
            flag = True
            break

    if flag:
        print("not prime")
    else:
        print("prime")

is_Prime(11)
```
