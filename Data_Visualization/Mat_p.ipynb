```python
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  

plt.plot(oscar_years,oscar_revenue)
```

```python
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275] 

plt.plot(oscar_years, oscar_revenue,"o-r",label='Oscar Winners')
plt.plot(non_oscar_years, non_oscar_revenue,"p--b",label = 'Non Oscar Winner')
plt.legend()
plt.title('years vs revenue')
plt.xlabel('years')
plt.ylabel('revenue')
```

```python
movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
revenue = [1005, 170, 427, 133, 232]

plt.bar(years,revenue,color='purple')

for i in range(len(revenue)):
    plt.text(years[i],revenue[i]+10,str(revenue[i]),ha='center')
```
