```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
bins=[0,30,70,90,100]
scores=np.random.normal(70,10,100)
#plt.hist(scores,color='skyblue',edgecolor='black',bins=5)
plt.hist(scores,color='skyblue',edgecolor='black',bins=bins)
plt.xlabel('scores')
plt.ylabel('no. of students')
```
