```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1.
np.random.seed(0)
data = np.random.normal(loc=70, scale=10, size=200)

plt.hist(data,bins=10,edgecolor='black',color='green')
plt.xlabel('Scores')
plt.ylabel('No. of students')
plt.title('Distrubution of scores of Students')

plt.axvline(
    x = data.mean(),
    label = 'mean value',
    color = 'blue',
    linestyle = '--',
    linewidth = 2
)
```

```python
# 2.
sns.get_dataset_names()
penguin = sns.load_dataset('penguins')
penguin.head()
```

```python
sns.relplot(
    data = penguin,
    x = 'flipper_length_mm',
    y = 'body_mass_g',
    hue ='species',
    style='species',
    size ='body_mass_g'
)
```

```python
# 3.

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.boxplot(
    data = tips,
    x = 'day',
    y = 'total_bill',
    hue = 'sex',
    legend=False

)
# “Sunday shows the highest median total bill, indicating higher typical spending. 
# Outliers are more frequent on weekends, suggesting occasional very high-value transactions.”
```

```python
# 4.
tips.head()
```

```python
ax= sns.scatterplot(
    data =tips,
    x = 'total_bill',
    y = 'tip',
    hue = 'sex',
    style = 'smoker'

)

ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# 5.
tips.head()
```

```python
tips_avg = tips.groupby(['day','time'], as_index=False)['tip'].mean()
tips_pivot = tips_avg.pivot(index='day', columns='time', values='tip')


ax=sns.heatmap(
    tips_pivot,
    annot =True,
    cmap="OrRd",
     fmt=".2f",
)
ax.set_title('Avergae Tip by Day and Time')
ax.set_xlabel('Time')
ax.set_ylabel('Day')
```
