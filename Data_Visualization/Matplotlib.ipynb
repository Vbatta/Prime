```python
# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.plot(years,oscar_revenue)
plt.title('Oscar Movie Revenue in Each Year')
plt.xlabel('Years')
plt.ylabel('Revenue (in $M) ')
```

```python
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

with plt.xkcd():
    plt.plot(years, oscar_revenue,color ='red',marker= "<",linestyle="-.",linewidth=3,label ='Oscar Movies')
    plt.plot(years, non_oscar_revenue,"<-y",label ='Non Oscar Movies')
    # "o--r"
    plt.title("Oscar Movies vs Non-Oscar Movies")
    plt.xlabel('Year')
    plt.ylabel('Revenue (in $M)')
    
    plt.legend()
    #plt.legend(loc = "upper left")
    print(plt.style.available)
    #plt.style.use("dark_background")
    plt.style.use("default")
    plt.grid()
    plt.show()
```

```python
# Scatter Plots

people = ["Person A", "Person B", "Person C", "Person D", "Person E",
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

colors = ['green' if x< 135 else 'red' for x in bp]
#plt.scatter(age,bp ,color =colors)
plt.scatter(age,bp ,s=bp,cmap = 'OrRd',c=bp)
plt.title('Age vs BP')
plt.xlabel('Age')
plt.ylabel('Blood Pressure')
plt.grid()

#plt.colorbar()
plt.colorbar(label='BP')

for i in range(len(people)):
    plt.annotate(people[i],xy=(age[i],bp[i]),xytext=(age[i]+1,bp[i]+1))

plt.xlim(min(age),max(age)+10)
plt.ylim(min(bp),max(bp)+5)
```

```python
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

explode=[0,0,0,0.3,0]
plt.style.use('default')
plt.pie(amounts,labels=expenses,autopct="%1.1f%%",wedgeprops={
    "edgecolor":"black",
    "linewidth":1,
    "linestyle":"--"},
    explode =explode,
    startangle =360,
    shadow= True
)

plt.show()
```
