```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
from sklearn.preprocessing import StandardScaler
```

```python
house_price = pd.read_csv('HousePricePrediction.csv')
house_price.head(5)
house_price.dropna()
house_price.isnull().sum()
house_price.head(5)
house_price = house_price.dropna()
```

```python
X = house_price.drop('SalePrice', axis=1)
y = house_price['SalePrice'].fillna(house_price['SalePrice'].mean)


cols = ['MSZoning', 'LotConfig', 'BldgType', 'Exterior1st']

X = pd.get_dummies(
    X,
    columns=cols,
    drop_first=True,
    dtype=int
)

```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)

X_train.isnull().sum()
y_train.isnull().sum()
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
print("R2 Score:", r2_score(y_test, y_pred))
print("mean absolute error",mean_absolute_error(y_test,y_pred))
print("mean square error",np.sqrt(y_test,y_pred))
print(mean_absolute_percentage_error(y_test,y_pred))
```

```python
# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model =LinearRegression()
model.fit(X_train_scaled ,X_test_scaled)
```
