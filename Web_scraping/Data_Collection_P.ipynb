```python
import requests
from bs4 import BeautifulSoup
```

```python
URL ='https://www.scrapethissite.com/pages/simple/'
res =requests.get(URL)

if res.status_code ==200:
    print("yes")
```

```python
with open("s_data/d.html","w", encoding="utf-8") as f:
    f.write(res.text)

with open("s_data/d.html","r", encoding="utf-8") as f:
    html_content =f.read()
soup = BeautifulSoup(html_content,"lxml")

all_h3 = soup.find_all('h3')
for h3 in all_h3:
    name =h3.get_text(strip=True)
    capital =h3.find_next('div').select_one('span.country-capital ').get_text(strip=True)
    print(name)
```
