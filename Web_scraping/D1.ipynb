```python
import requests
from bs4 import BeautifulSoup
```

```python
page_count =1

while True:
    URL =f"https://quotes.toscrape.com/page/{page_count}/"
    res= requests.get(URL)

    soup= BeautifulSoup(res.text,"lxml")
    quotes =soup.select("div.quote")

    if not quotes:
        print("no valid pages anymore")
        break

    with open(f"scraped_data1/quotes{page_count}.html","w",encoding="utf-8")as f:
        f.write(res.text)
        print(f"page downloaded {page_count}")

    page_count +=1
```

```python

with open("scraped_data1/quotes1.html","r",encoding="utf-8")as f:
    html_content = f.read()

soup = BeautifulSoup(html_content,"lxml")

all_quotes = soup.select("div.quote")
life_quotes =[]

for q in all_quotes:
    all_tags = []
    for tag in q.select(".tags .tag"):
        all_tags.append(tag.get_text())
        #print(all_tags)

    if 'life' in all_tags:
        text =q.select_one("span.text").get_text()
        author = q.select_one("small.author").get_text()
        life_quotes.append([text,author])

print(life_quotes)
```

```python
page_count = 1

while True:
    try:
        with open(f"scraped_data1/quotes{page_count}.html","r",encoding="utf-8") as f:
            html_content = f.read()
    except FileNotFoundError:
        print("No more saved pages")
        break

    soup = BeautifulSoup(html_content,"lxml")

    all_quotes = soup.select("div.quote")
    life_quotes = []

    for q in all_quotes:
        all_tags = []
        for tag in q.select(".tags .tag"):
            all_tags.append(tag.get_text())

        if 'life' in all_tags:
            text = q.select_one("span.text").get_text()
            author = q.select_one("small.author").get_text()
            life_quotes.append([text, author])

    print(life_quotes)

    page_count += 1
```
