```python
from bs4 import BeautifulSoup

with open("scraped_data/data1.html", "r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content,"lxml")
```

```python
#soup
#soup.find('h1')
#soup.h1
all_h3 = soup.find_all('h3')
#print(all_h3)
type(all_h3)

for h3 in all_h3:
    #print(h3.get_text())
    name = h3.get_text(strip = True)
    #print(h3.find_parent('div'))
    #print(h3.find_parent('div')['class'])
    #print(h3.find_next('div'))
    #print(h3.find_next('div').select('span.country-population')[0].get_text(strip =True))
    population = h3.find_next('div').select_one('span.country-population').get_text(strip =True)
    print(population)
    print(h3.attrs)
    #print(h3.parent)
    #print(h3.child)
```

```python

all_h3 = soup.find_all('h3')
all_countries =[]
for h3 in all_h3:
    
    name = h3.get_text(strip = True)

    population = h3.find_next('div').select_one('span.country-population').get_text(strip = True)

    all_countries.append([name,population])
```

```python
all_countries
```

```python
import pandas as pd
df =pd.DataFrame(all_countries,columns =['Name','Population'])
df
```

```python
df.to_csv("cleaned_data/data.csv",index =False)
```
