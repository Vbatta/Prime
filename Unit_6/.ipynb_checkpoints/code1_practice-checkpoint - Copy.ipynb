```python
import json
```

```python
def filename(file):
    with open(file,"r")as f:
       data = json.load(f)
    return data
```

```python
data=filename('store.json')
data
```

```python
def clean_data(data):
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5}
    cleaned_data = []
    unique_users = set()
    
    # clean rating
    for user in data:
        raw_rating = user['rating']
        if type(raw_rating) == str:
            raw_rating= raw_rating.strip().lower()
            if raw_rating in text_to_num:
                user['rating']=text_to_num[raw_rating]
            print(raw_rating)

    # handle missing age
        raw_age =user.get('age')
        if(raw_age == None):
            user['age'] = None
            
     # deduplication
        if(user['name']  in unique_users):
            continue
        unique_users.add(user['name'])
        cleaned_data.append(user)
    return cleaned_data
```

```python
data1 =clean_data(data)
data1
```

```python
# Get meaningful insights from data

def get_insights(data1):
    tot_rating =0
    # avg_rating
    for user in data1:
        tot_rating+= float(user['rating'])
    
    print(f"average rating is {tot_rating/len(data1)}")

    # percentage of users with poor rating.
    poor_ratings = 0
    for user in data1:
        if float(user['rating'])<3:
            poor_ratings +=1
    print(f"percentage of poor ratings {poor_ratings/len(data) *100}")
```

```python
get_insights(data1)
```
