```python
import json
```

```python
def load_data(filename):
    with open(filename,"r")as f:
        data =json.load(f)

    return data
```

```python
data=load_data("store.json")
print(data)
```

```python
# Clean & structure the data
def clean_data(data):
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5}
    cleaned_data =[]
    unique_user =set()

    for user in data:
        # Clean ratings
        raw_rating = user["rating"]
        if type(raw_rating) == str:
            raw_rating = user["rating"].strip().lower()
            if raw_rating in text_to_num:
                raw_rating = text_to_num[raw_rating]
            user["rating"] = raw_rating
        print( user["rating"])

        if(user.get("age")==None):
            user["age"]=None

        if(user["name"] in unique_user):
            continue

        unique_user.add(user["name"])
        cleaned_data.append(user)

    print(cleaned_data)
    return cleaned_data
    
       
        
```

```python
data = clean_data(data)
```

```python
def get_insights(data):
    tot_rating = 0
    for user in data:
        tot_rating += float(user["rating"])
        
    print(f"total rating is : {tot_rating/len(data)}")

    count =0
    for user in data:
        if(float(user["rating"])<=3):
            count+=1

    print(f"percentage of users : {(count/len(data))*100 }%")
```

```python
get_insights(data)
```

```python
def get_rec(data):
    recommendations=[]
    for user in data:
        curr_rec ={}
        curr_rec["name"]=user["name"]
        if(float(user["rating"])>=4):
            curr_rec["brand"]="apple"
        else:
             curr_rec["brand"]="samsung"
        recommendations.append(curr_rec)
    return recommendations
```
