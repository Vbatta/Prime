```python
df = pd.read_csv("globalAirQuality.csv")
df.head(2)
```

```python
# Indexing & Data Selection

# Selecting Columns
df['city']
df[['city','aqi']]

# Selecting Rows 
# Index label
df.loc[2] # row 3rd
df.loc[0:2] # start idx: end idx (inclusive) , row0 to row3
df.loc[0,"country"] # 0th row & specific col

# Index Position 
df.iloc[2] # row 3rd
df.iloc[0:2] # start idx: end idx (exclusive) , row0 to row2
df.iloc[0,2] # 0th row & 3rd col

df.loc[0:2,["city","aqi"]]
# df.iloc[0:2,["city","aqi"]] , would give error


# Selecting single Cell (at ,iat for faster access)

df.at[0,"city"]
df.iat[0,2]
```

```python
# <-- Important Practice -->
# View , does permanent change , gives warning
# cities = df['city'] 
# cities[0] ="Mumbai"
# cities[0] ="New York"

# copy , good practice
cities = df['city'].copy()
cities[0] ="Mumbai"
cities # copy
df # original
```

```python
# Filtering of Data (Important)
df[df['aqi']>100]
df[(df['aqi']>100) & (df['temperature']> 30)]

aqi_data = df[(df['aqi']>100) & (df['temperature']> 30)][["city","aqi"]]
aqi_data

# Important
aqi_data.iloc[6] # Position based
aqi_data.loc[6] # Index based
```
