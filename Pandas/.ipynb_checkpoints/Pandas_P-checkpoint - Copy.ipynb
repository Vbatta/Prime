```python
import pandas as pd
df = pd.read_csv("raw_data.csv")
df.head(3)
```

```python
df1 =df.copy()
# Apply 
df1['tax'] =df1['income'].apply(lambda x: "20%" if x >60_000 else "10%")
df1

# Map
Map_gender ={'Male':'M','Female':'F','Unknown':'U'}
df1['gender'] =df1['gender'].map(Map_gender)
df1

# Assign
df1= df1.assign(new_income = df1['income']*1.2)
df1

# Replace
df1["country"] =df1["country"].replace("USA","US")
df1.head(3)
```

```python
# Renaming columns
df1.columns = ["Id","Name","Age","Columns","Gender","Income","Tax","New_Income"]
df1

df1.rename(columns ={'Id':'id'})
df1.rename(index ={1:'First'})

# Sort values
df1.sort_values("Income")
df1.sort_values("Income",ascending = False)
sorted_df =df1.sort_values(['Age','Income'])
sorted_df

# Sort Index
sorted_df.sort_index()
# Reset Index
sorted_df.reset_index()
```

```python
sorted_df.reset_index(drop =True)
```

```python
# Ranking
sorted_df['Ranking'] =sorted_df['Income'].rank(ascending =False)
sorted_df
sorted_df['Ranking'] =sorted_df['Income'].rank(ascending =False,method ="dense")
sorted_df
```

```python
#sorted_df[["Name","Age","Columns","Income","Tax","New_Income","Ranking","Id"]]
# Generate new column order using List Comprehensions.
df2 =df.copy()
new_col_ord = [x for x in df2.columns if x != 'id']+["id"]
new_col_ord 
df2[new_col_ord]
```

```python
# Write output to csv after cleaning data

df2 =df.copy()
df2 =df2.fillna(0)
df2 =df2.drop_duplicates()
df2 =df2.sort_values("income")
df2 = df2.sort_index()
df2 =df2.reset_index(drop=True)
df2
```

```python
# Grouping & Aggregation Method
df2 =df.copy()
df2.groupby("country")["income"].mean()
df2.groupby("country")["income"].min()
df2.groupby("country")["income"].max()

df2.groupby("country")["income"].agg(["mean","max","min"])
df2.groupby("country")["income"].aggregate(["mean","max","min"])
df2.groupby("country")["income"].aggregate(avg_sal="mean",max_sal="max",min_sal="min")

df2.groupby("country").agg({
    "income":"max",
    "age" : "min"
})

df2.groupby("country").agg(
  max_sal = ("income","max"),
  min_age = ("age", "min")
)
```
