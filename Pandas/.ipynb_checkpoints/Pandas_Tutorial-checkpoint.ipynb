```python
df2 =df.copy()
df2['tax'] =df2['income'].apply(lambda x: '20%' if x >=60_000 else '10%')
df2

gender_map ={"Male":'M' , 'Female' :'F','Unknown':'U'}
df2['gender'] =df2['gender'].map(gender_map)
df2

df2 =df2.assign(new_income = df2['income'] * 1.1)
df2

df2['country'] =df2['country'].replace('USA','US')
df2
```

```python
df2.columns=['Id','Name','Age','Country','Gender','Income','Tax','New_Income']
df2

df2.rename(columns={'Income': 'Salary'})
df2
#
df2=df2.rename(columns={'Salary':'Income'})

df2.rename(index={1:'First'})
df2

df2.sort_values('Income',ascending = False)
df2['Income'].sort_values()
df2 =df2.fillna(50)
sorted_df=df2.sort_values(["Income","Age"]) # first arrange in order of Income then Age
sorted_df.sort_index() # Imp.

sorted_df.reset_index() 
sorted_df.reset_index(drop = True) 

sorted_df["Ranking"]=sorted_df["Income"].rank(ascending =False ,method = "dense")
sorted_df
```

```python
df.groupby("country")["income"].mean()
df.groupby("country")["income"].min()
df.groupby("country")["income"].max()

df.groupby("gender")["income"].max()

df.groupby("country")["income"].agg(["mean","max","min"])
df.groupby("country")["income"].aggregate(["mean","max","min"])

df.groupby("country")["income"].aggregate(avg_salary="mean",max_sal="max",min_sal="min")

df.groupby("country").agg({
    "income":"max",
    "age":"min" 
})

df.groupby("country").agg(
    max_salary =("income","max"),
    min_age =("age","min")
)
```

```python
df=pd.DataFrame({
"country":["USA","USA","India","India"],
"year":[2020,2021,2020,2021],
"sales":[100,120,90,110],
"profit":[20,25,18,22]
})

melted_df=df.melt(
    id_vars =["country","year"],
    value_vars=["sales","profit"],
    var_name="metrics",
    value_name ="value"
)

melted_df

melted_df.pivot(
    index=["country","year"],
    columns="metrics",
    values="value"
)
```

```python
# Merging and joining Data
df_customers =pd.DataFrame({
"customer_id":[1,2,3,4],
"name":["Adam","Bob","Charlie","Dave"]
})

df_orders = pd.DataFrame({
"order_id":[101,102,103,104],
"customer_id":[2,1,4,5],
"amount":[250,120,300,180]
})

df_customers
```

```python
df_orders
```

```python
pd.merge(df_customers,df_orders ,on="customer_id")
pd.merge(df_customers,df_orders ,on="customer_id",how="left")
pd.merge(df_customers,df_orders ,on="customer_id",how="right")
pd.merge(df_customers,df_orders ,on="customer_id",how="outer")
```
